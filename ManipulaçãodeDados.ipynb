```python
import pandas as pd

dados = {
    'Produto':['A','B','C','D'],
    'Valor ':[50, 120, 180, 250],
    'Quantidade':[10, 100, 1000, 500]
    }

df = pd.DataFrame(dados)

filtro = df[df['Valor ']>150]
print("\nLinhas com valor > 150 :")
filtro  ##trocando em miúdos print valores maiores que 150
```

```python
filtro2= df[(df['Valor ' ]>200) & (df['Quantidade']>400)]
print("\nLinhas com valor maior que 150 e Quantidade maior que 400: ")
filtro2

##trocando em miúdos apresentar o print que possui duas condições em que o valor é maior que 150 e a quantidade maior que 400
```

```python

filtro3=df[df['Produto'].isin(['A','C'])]
print("\nLinhas com Produto A ou C : ")
filtro3 ##Trocando em miúdos verifica-se se as strings apresentadas estão dentro do dataframe
```

```python
filtro4 =df[df['Valor '].between(100,150)]
print("\nLinhas com Valor entre 100 e 150:")
filtro4 ##O between verifica se existe linha entre 1 valor e outro
```

```python
filtro5 = df[df['Produto'].str.contains('^A')]
print("\nLinhas com Produto iniciado por 'A':")
filtro5
```
